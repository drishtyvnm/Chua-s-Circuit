```python
from matplotlib import pyplot as plt
from matplotlib.animation import ArtistAnimation
import numpy as np
from scipy.integrate import odeint
```

Chua's circuit was invented in 1983 by **Leon O. Chua**. Chua's circuit is a simple electronic circuit that exhibits classic chaotic behavior which means roughly that it is a *nonperiodic oscillator*. The ease of construction of the circuit has made it a ubiquitous real-world example of a chaotic system, leading some to declare it "a paradigm for chaos".

```python
fig=plt.figure()
m0 = -1/7
m1 = 2/7
tmax = 18;
def chua(x, t):
  return [a * (x[1] - (m1 *x[0] + (m0 - m1) /2 *
   (np.abs(x[0] + 1) - \
    np.abs(x[0] - 1)))), x[0] -x[1] + x[2], -15 * x[1]]
```

The dynamics of *Chua's circuit* can be accurately modeled by means of a system of three nonlinear ordinary differential equations in the variables $x(t)$, $y(t)$, and $z(t)$, which represent the voltages across the capacitors $C1$ and $C2$ and the electric current in the inductor $L1$ respectively. These equations are:

$$\frac{dx}{dt}=α[y-x-f(x)]$$
$$RC_2\frac{dy}{dt}=x-y+Rz$$
$$\frac{dz}{dt}=-βy$$
